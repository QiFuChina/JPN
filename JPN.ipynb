#### Straight lines
All straight lines can be expressed in the form  y=mx+cy=mx+c . 
The number  mm  is the slope of the line. The slope is how much  yy  increases by when  xx  is increased by 1.0. 
The number  cc  is the y-intercept of the line. 
It's the value of  yy  when  xx  is 0.

#### Fitting the model
To fit a straight line to the data, we just must pick values for $m$ and $c$.
These are called the parameters of the model, and we want to pick the best values possible for the parameters.
That is, the best parameter values *given* the data observed.
Below we show various lines plotted over the data, with different values for $m$ and $c$.

#### Calculating the cost
The best line is the one that minimises the following calculated value.

$$ \sum_i (y_i - mx_i - c)^2 $$
`cost = lambda m,c: np.sum([(d[i] - m * w[i] - c)**2 for i in range(w.size)])`

#### Minimising the cost
We want to calculate values of  mm  and  cc  that give the lowest value for the cost value above. For our given data set we can plot the cost value/function. Recall that the cost is:
$$ Cost(m, c) = \sum_i (y_i - mx_i - c)^2 $$

### Gradient descent
In gradient descent, we select a random guess of a parameter and iteratively improve that guess.
For instance, we might pick $1.0$ as our initial guess for $m$ and then create a `for` loop to iteratively improve the value of $m$.
The way we improve $m$ is to first take the partial derivative of our cost function with respect to $m$.

### Cost function
Recall that our cost function for simple linear regression is:

$$
Cost(m, c) = \sum_i (y_i - mx_i - c)^2
$$

### Calculate the partial derivatives
We calculate the partial derivative of $Cost$ with respect to $m$ while treating $c$ as a constant.
Note that the $x_i$ and $y_i$ values are all just constants.
We'll also calculate the partial derivative with respect to $c$ here.

$$
\begin{align}
Cost(m, c)                       &= \sum_i (y_i - mx_i - c)^2      \\[1cm]
\frac{\partial Cost}{\partial m} &= \sum 2(y_i - m x_i -c)(-x_i) \\
                                 &= -2 \sum x_i (y_i - m x_i -c) \\[0.5cm]
\frac{\partial Cost}{\partial c} & = \sum 2(y_i - m x_i -c)(-1) \\
                                 & = -2 \sum (y_i - m x_i -c)   \\
\end{align}
$$

### Code the partial derivatives

```python
def grad_m(x, y, m, c):
  return -2.0 * np.sum(x * (y - m * x - c))
def grad_c(x, y, m , c):
  return -2.0 * np.sum(y - m * x - c)
```
